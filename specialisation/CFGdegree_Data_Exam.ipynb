1. PANDAS (25 points)

```python
# 1 (2 marks). Load the netflix_titles.csv dataset as a Pandas dataframe:

import pandas as pd # import pandas as pd first

df = pd.read_csv('netflix_titles.csv') # read in using read_csv
```

```python
# 2 (2 marks). Show the first 10 rows of the dataset

df.head(10) # use head combined with agurment 10 to show first 10 rows
```

```python
# 3 (1 mark). Show the last 5 rows

df.tail(5) # use tail instead with 5
```

```python
# 4 (4 marks). Count the number of rows which have no missing value in any of the columns, and show the resulting data frame.

df_no_NA = df.dropna() # drop rows which have missing values in any column; axis 0 (rows) is default
count = df_no_NA.shape[0] # count by using shape[0] (shape returns n row, n col so 0 = n row)
df_no_NA # show data frame
```

```python
# 5 (3 marks). Drop everything that is not a 'Movie' and show the resulting data frame. Use any method for this.

df_movies = df[df['type'] == 'Movie'].copy() # by selecting rows where type is movie, this drops everything that is not a movie
# use copy to create a deep copy rather than view of slice
df_movies # show data frame with only movies
```

```python
# 6 (7 marks). For every movie where the duration is missing, replace it with the average
  # duration of the movies in the dataset and return the resulting data frame

# before we can do anything, need to format duration column so it is numeric (remove min str first and then convert)
df_movies['duration'] = df_movies['duration'].str.replace(' min', '') # use replace to remove ' min' with ''
# df_movies.head(5) to check it's worked

df_movies['duration'] = df_movies['duration'].astype(float) # convert to float for calculating mean

df_movies['duration_imputed_mean'] = df_movies['duration'].fillna(df_movies['duration'].mean()) # impute missing with mean
df_movies['duration_imputed_median'] = df_movies['duration'].fillna(df_movies['duration'].median()) # impute missing with median (putting both as question didn't specify which average)

# df_movies[['duration','duration_imputed_mean', 'duration_imputed_median']][df_movies['duration'].isna()] # to sanity check what you have changed

# return the resulting data frame
df_movies
```

```python
# 7 (3 marks). Are there any duplicates in the data? If yes, please remove them.

# first check if there are duplicates, then remove if there are (2 steps as specified in question)
dup = df_movies[df_movies.duplicated()] # select duplicated data
n_dup = dup.shape[0] # use shape[0] to check n rows
n_dup # as this is zereo, there are no duplicates so no need to remove them

# if there had been duplicates, you could use df_movies.drop_duplicates() to remove them / just used that first and then checked shape to see if anything changed
```

```python
# 8 (2 marks). Rename all columns to start with a capital letter.
df_movies.columns = [x.title() for x in df_movies.columns] # use title function to make each column start with capital letter
# df_movies.columns # to check it's worked
```

```python
# 9 (1 mark). Save the cleaned dataset as a new csv file.
df_movies.to_csv('netflix_movies_cleaned.csv', index = False) # use index = false to stop extra column
```

```python
# 1 (10 marks). Create a visualisation (by using the plot that is most appropriate for this) which shows the frequency of movies by year of release.

import matplotlib.pyplot as plt # import matplotlib.pyplot

# read cleaned data back in using read_csv
df_clean = pd.read_csv('netflix_movies_cleaned.csv')

# for vis, want year of release on x-axis (Release_Year), frequency on y-axis
# need to calculate frequency first using value_counts and reset the index so returns a data frame with two columns for plotting
freq_df = df_clean['Release_Year'].value_counts().reset_index()
# freq_df # to check it's worked as intended

# make year numeric for nicer spacing of x-ticks below
freq_df['Release_Year'] = freq_df['Release_Year'].astype(int)

plt.figure(figsize=(20, 5)) # create figure object and set dimensions of width and height
plt.bar(freq_df['Release_Year'], freq_df['count'])  # use the two columns in freq_df to plot frequency on y and release year on x
plt.xlabel('Release Year') # change x label title
plt.ylabel('Frequency') # change y label title
plt.title('Frequency of Movies By Release Year') # set overall title
# this makes sure all years are plotted but are also evenly spaced (so if there are no values it will still plot on x)
# provide min and max years and make all values using range, use rotation to make them fit nicely
plt.xticks(range(min(freq_df['Release_Year']), max(freq_df['Release_Year'])+1), rotation=45)
plt.tight_layout() # nicer spacing
plt.show() # show plot
```

```python
# 2 (15 marks). Create a pie chart which shows the 5 top most popular genres of the movies
  # in the dataset.

# read cleaned data back in to be safe
df_clean = pd.read_csv('netflix_movies_cleaned.csv')
# df_clean.columns # reminder of column names
# the genres are contained in 'Listed_In' but sepeerated by commas so need to deal with that first

# need to first separate out genres from Listed_In
# used solutions from https://stackoverflow.com/questions/12680754/split-explode-pandas-dataframe-string-entry-to-separate-rows
df_clean['Genre'] = df_clean['Listed_In'].str.split(',') # have to make each genre an item in a list first for explode to work properly
df_genres = df_clean.explode('Genre') # use explode to split values in list into seperate rows
# df_genres # check output is as intended (now loads more rows)

# calculate counts on genres using value counts (reset index as before to make nice dataframe)
genre_freq = df_genres['Genre'].value_counts().reset_index()

# select top 5 biggest counts using head
top_5 = genre_freq.head(5)
# top_5 # check

# plot pie chart using plt.pie
plt.figure(figsize=(8, 8)) # set figure object and dimensions first
plt.pie(top_5['count'], labels=top_5['Genre'], autopct='%1.1f%%') # give values in count first, labels from genre and use autopct to calculate percentages
plt.title('Pie Chart of Top 5 Most Popular Genre of Movies')
plt.show() # show chart
```

3. NUMPY (25 points)

```python
# 1 (2 marks). Running the code in this cell throws an error. Why do we get this error
  # and how do we fix it?

# You get this error because you have not imported numpy as 'np' before running. You need to first import numpy as np before running.
import numpy as np
# use this numpy array for all exercises in this section
arr = np.array([0, 56, 82, 500, 140, 43, 92, 10, 25, 11, 18, 26, 150, 240, 364, 196])
arr
```

```python
# 2 (5 marks). How many odd numbers do we have in the array? Show how we get this result using numpy

odd = arr[arr % 2 == 1] # return odd numbers in array using where remainder of division 2 is equal to zero
num_odd = len(odd) # cal length of odd
num_odd # present
```

```python
# 3 (5 marks). Replace all odd numbers in arr with 999
arr[arr % 2 == 1] = 999
arr
```

```python
# 4 (8 marks). The above is a 1D array. Convert it into a 4D array and reverse its columns.

# # Desired output
# array([[500, 82, 56, 0],
# [ 10, 92, 999, 140],
# [ 26, 18, 999, 999],
# [196, 364, 240, 150]])

# first reshape, 16 items = 4 x 4
reshaped_arr = arr.reshape(4,4)
reshaped_arr

# then reverse cols using flipr
reversed_arr = np.fliplr(reshaped_arr)
reversed_arr
```

```python
# 5 (5 marks). Find and return all values which are higher than 100 and lower or equal than 500.

new_arr = reversed_arr[(reversed_arr > 100) & (reversed_arr <= 500)] # both conditions need to be satisfied to be add to new_arr
new_arr
```
